```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
df = pd.read_json (r'Data/train/train.json')
```

```python
df['path'] = df['image'].map(lambda x: Path('../malaria/')/ x['pathname'][1:])
```

```python
df.head()
```

```python
train = pd.DataFrame(columns=('path','category','x1','x2','y1','y2'))
```

```python
for i in range(len(df)):
    path = df['path'][i]
    objects = df['objects'][i]
    for j in range(len(objects)):
        train=train.append(pd.DataFrame({'path':[path],
                                         'category':objects[j]['category'],
                                         'x1':objects[j]['bbx']['x1'],
                                         'x2':objects[j]['bbx']['x2'],
                                         'y1':objects[j]['bbx']['y1'],
                                         'y2':objects[j]['bbx']['y2']}),ignore_index=True)
```

```python
train.to_csv (r'Data/train/train_df.csv', index = None)
```
